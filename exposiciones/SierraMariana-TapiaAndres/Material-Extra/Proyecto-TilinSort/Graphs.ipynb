```python
import csv
import matplotlib.pyplot as plot

tiempo = []
iteracion = []
parallelTime = []
parallelTaskTime = []

with open('SerialExecuteTime.csv', 'r') as file:
    reader = csv.reader(file)
    next(reader)  # Saltar la primera fila (encabezado)
    for row in reader:
        Iteracion, Tiempo = map(float, row)  # Convertir de cadena a flotante
        tiempo.append(Tiempo)

with open('ParallelExecuteTime.csv', 'r') as file:
    reader = csv.reader(file)
    next(reader)  # Saltar la primera fila (encabezado)
    for row in reader:
        Iteracion, Tiempo = map(float, row)  # Convertir de cadena a flotante
        parallelTime.append(Tiempo)

with open('ParallelTaskExecuteTime.csv', 'r') as file:
    reader = csv.reader(file)
    next(reader)  # Saltar la primera fila (encabezado)
    for row in reader:
        Iteracion, Tiempo = map(float, row)  # Convertir de cadena a flotante
        iteracion.append(Iteracion)
        parallelTaskTime.append(Tiempo)

punto = (1500,0.01)

plot.scatter(*punto, color='black', marker='o', label='Punto de equilibrio')
plot.plot(iteracion, tiempo, color='blue', label="Serial")
plot.plot(iteracion, parallelTaskTime, color='purple', label='ParalelTaskTime')
plot.plot(iteracion, parallelTime, color='red', label='Paralel')
plot.title('Rendimiento temporal de TilinSort')
plot.xlabel('Iteracion')
plot.ylabel('Tiempo de ejecucion (segundos)')
plot.legend()
plot.show()
```
